```python
def imshow(dataset) :
    inp,_ = next(iter(torch.utils.data.DataLoader(dataset,batch_size=64,shuffle=True)))
    inp = torchvision.utils.make_grid(inp)
    plt.imshow(inp.permute(1, 2,0))
    plt.show()
imshow(mnist_train)
imshow(fmnist_train)
imshow(emnist_train)
```

```python
import torch.optim as optim
def train(net,lr,wd,dataset,loss_fn) :
    net.train()
    optimizer = optim.SGD(net.parameters(),lr=lr,nesterov=True,momentum=0.9)
    for x,y in dataset :
        x = x.cuda()
        y = y.cuda()
        optimizer.zero_grad()
        y_pred = net(x)
        loss_fn(y_pred,y).backward()
        L2 = wd*torch.cat([x.view(-1) for x in net.fc.parameters()]).pow(2).sum()
        L2.backward()
        optimizer.step()

def inference(net,dataset,loss_fn) :
    loss = 0
    with torch.no_grad() :
        net.eval()
        for x,y in dataset :
            x = x.cuda()
            y = y.cuda()
            y_pred = net(x)
            loss += loss_fn(y_pred,y).detach().cpu().numpy().item()
    return loss/dataset.dataset.__len__()

def run_epoch(net,lr,batch_size,train_dataset,val_dataset,test_dataset,train_loss,val_loss,max_epochs,wd=0) :
    train_losses = []
    val_losses = []
    test_losses = []
    train_loader = torch.utils.data.DataLoader(train_dataset,shuffle=True,batch_size=batch_size)
    val_loader = torch.utils.data.DataLoader(val_dataset,shuffle=True,batch_size=batch_size)
    test_loader = torch.utils.data.DataLoader(test_dataset,shuffle=True,batch_size=batch_size)
    
    for i in range(max_epochs+1) :
        train(net,lr,wd,train_loader,train_loss)
        if i%5 == 0:
            train_losses.append(inference(net,train_loader,val_loss))
            val_losses.append(inference(net,val_loader,val_loss))
            test_losses.append(inference(net,test_loader,val_loss))
            print("epoch : {},train_loss : {:.2f},val_loss : {:.2f},test_loss : {:.2f}".format(i,train_losses[-1],val_losses[-1],test_losses[-1]))
    return train_losses,val_losses,test_losses
       
```

We will build a network with 2 convolutional layers and 1 fully connected layer to test the performance of the various convolutional blocks.
The activations used in the convolutional layer would be relu
All pooling operations would be max pool

```python
class MNISTNet(nn.Module) :
    def __init__(self,block_type,init_filter=32,num_classes=10) :
        super(MNISTNet,self).__init__()
        self.initial_conv = nn.Sequential(nn.Conv2d(1,init_filter,3,padding=1),nn.ReLU())
        self.layer_1 = nn.Sequential(block_type(init_filter,2*init_filter,3,padding=1),nn.MaxPool2d(2,2))
        self.layer_2 = nn.Sequential(block_type(2*init_filter,4*init_filter,3,padding=1),nn.MaxPool2d(2,2))
        self.fc = nn.Linear(4*init_filter*49,num_classes)
    def forward(self,x) :
        x = self.initial_conv(x)
        x = self.layer_1(x)
        x = self.layer_2(x)
        x = x.view(x.shape[0],-1)
        x = self.fc(x)
        return x
```

```python
class SimpleConv(nn.Module) :
    def __init__(self,init_filters,final_filters,kernel,padding=0,activation=nn.ReLU) :
        super(SimpleConv,self).__init__()
        self.conv = nn.Conv2d(init_filters,final_filters,kernel,padding=padding)
        self.activation = activation()
    def forward(self,x) :
        return self.activation(self.conv(x))
```

```python
class DoubleConv(nn.Module) :
    def __init__(self,init_filters,final_filters,kernel,padding=0,activation=nn.ReLU) :
        super(DoubleConv,self).__init__()
        self.conv1 = nn.Sequential(nn.Conv2d(init_filters,final_filters,kernel,padding=padding),activation())
        self.conv2 = nn.Sequential(nn.Conv2d(final_filters,final_filters,kernel,padding=padding),activation())
    def forward(self,x) :
        return self.conv2(self.conv1(x))
```

```python
class InceptionConv(nn.Module) :
    def __init__(self,init_filters,final_filters,kernel,padding=0,activation=nn.ReLU) :
        super(InceptionConv,self).__init__()
        self.conv1 = nn.Sequential(nn.Conv2d(init_filters,final_filters/8,1,padding=0),activation())
        self.conv2 = nn.Sequential(nn.Conv2d(init_filters,final_filters/8,1,padding=0),activation(),nn.Conv2d(final_filters/8,final_filters/4,3,padding=1),activation())
        self.conv3 = nn.Sequential(nn.Conv2d(init_filters,final_filters/8,1,padding=0),activation(),nn.Conv2d(final_filters/8,final_filters/2,5,padding=2),activation())
        self.conv4 = nn.Sequential(nn.MaxPool2d(3,stride=1,padding=1),nn.Conv2d(init_filters,final_filters/8,3,padding=1),activation())

    def forward(self,x) :
        x1 = self.conv1(x)
        x2 = self.conv2(x)
        x3 = self.conv3(x)
        x4 = self.conv4(x)
        return torch.cat([x1,x2,x3,x4],dim=1)
```

```python
def zero_one_loss(y_pred,y_true) :
    indices = torch.argmax(y_pred,dim=1,keepdim=True).type_as(y_true)
    y_true = y_true.view(-1,1)
    return torch.sum(torch.eq(indices-y_true,0))
```

```python
mnist_net = MNISTNet(SimpleConv).cuda()
train_loss = nn.CrossEntropyLoss()
val_loss = zero_one_loss
mtr,mv,mte = run_epoch(mnist_net,1e-3,512,mnist_train,mnist_val,mnist_test,train_loss,val_loss,250,0)
plt.plot(mtr)
plt.plot(mv)
plt.plot(mte)
plt.legend(['training','validation','test'],loc='lower right')
plt.show()
torch.save(mnist_net.state_dict(),'mnist_net_simpleconv.mod')
```

```python
fmnist_net = MNISTNet(SimpleConv).cuda()
train_loss = nn.CrossEntropyLoss()
val_loss = zero_one_loss
ftr,fv,fte = run_epoch(fmnist_net,1e-3,512,fmnist_train,fmnist_val,fmnist_test,train_loss,val_loss,250,0)
plt.plot(ftr)
plt.plot(fv)
plt.plot(fte)
plt.legend(['training','validation','test'],loc='lower right')
plt.show()
torch.save(fmnist_net.state_dict(),'fmnist_net_simpleconv.mod')
```

```python
torch.cuda.empty_cache()
emnist_net = MNISTNet(SimpleConv,num_classes=47).cuda()
train_loss = nn.CrossEntropyLoss()
val_loss = zero_one_loss
etr,ev,ete = run_epoch(emnist_net,1e-3,512,emnist_train,emnist_val,emnist_test,train_loss,val_loss,250,0)
plt.plot(etr)
plt.plot(ev)
plt.plot(ete)
plt.legend(['training','validation','test'],loc='lower right')
plt.show()
torch.save(emnist_net.state_dict(),'emnist_net_simpleconv.mod')
```
